This rudimentary example looks correct. We have

$$B_z = \frac{1}{\sqrt{2}},\quad B_x = \frac{1}{\sqrt{2}}, \quad B_y = 0$$

after the rotation by $\frac{\pi}{4}$ about the $y$-axis

## A step further

\begin{align} 
G_{1,-2}    &=\partial_{y} B_{x}=\partial_{x} B_{y}  \label{eq:G1n2}\\
G_{1,-1}    &=\partial_{y} B_{z}=\partial_{z} B_{y} \\ 
G_{1,0}     &=\partial_{z} B_{z}=-\partial_{x} B_{x}-\partial_{y} B_{y} \\ 
G_{1,1}     &=\partial_{x} B_{z}=\partial_{z} B_{x} \\ 
G_{1,2}     &=\frac{1}{2}\left(\partial_{x} B_{x}-\partial_{y} B_{y}\right) 
\end{align}

Now I'll add a gradient in $B_z$ (and in $B_x$ and $B_y$ as well) by making $G_{1,0}$ non-zero. This time, instead of using the flat x-y plane, I give the following argument to PENTrack:

```bash
BCutPlane 0 0 0  1 0 0  0 1 1  100 100
```

````python
f = root_open('/ucn/orithyia_data/jberean2/PENTrack/test/HarmonicFieldTest/B0z_G10.out.root')
# f = root_open('BFCut.out.root')

x = np.empty((0,1), float)
y = np.empty((0,1), float)
z = np.empty((0,1), float)
Bx= np.empty((0,1), float)
By= np.empty((0,1), float)
Bz= np.empty((0,1), float)
# canvas = Canvas(width=700, height=500)
# hsurf = Hist2D(10,0,1,1000,0,1);
for evt in f.mytree:
    x  = np.append(x, evt.x)
    y  = np.append(y, evt.y)
    z  = np.append(z, evt.z)
    Bx = np.append(Bx, evt.Bx)
    By = np.append(By, evt.By)
    Bz = np.append(Bz, evt.Bz)
#     print(evt.Bz)
# hsurf.Draw("SURF");
# canvas.Draw()

fig, ax = plt.subplots()
ax = plt.axes(projection='3d')
ax.scatter(x, y, Bz, c=Bz, cmap='viridis', linewidth=0.5);
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$B_z$');
fig.tight_layout()
````
