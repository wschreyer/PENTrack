# Fields with Uniform Gradients

\begin{align} 
G_{1,-2}    &=\partial_{y} B_{x}=\partial_{x} B_{y}  \label{eq:G1n2}\\
G_{1,-1}    &=\partial_{y} B_{z}=\partial_{z} B_{y} \\ 
G_{1,0}     &=\partial_{z} B_{z}=-\partial_{x} B_{x}-\partial_{y} B_{y} \\ 
G_{1,1}     &=\partial_{x} B_{z}=\partial_{z} B_{x} \\ 
G_{1,2}     &=\frac{1}{2}\left(\partial_{x} B_{x}-\partial_{y} B_{y}\right) 
\end{align}

Now I'll add a gradient in $B_z$ (and in $B_x$ and $B_y$ as well) by making $G_{1,0}$ non-zero. This time, instead of using the flat x-y plane, I give the following argument to PENTrack:

```bash
BCutPlane 0 0 0  1 0 0  0 1 1  100 100
```

````python
f = root_open('/ucn/orithyia_data/jberean2/PENTrack/test/HarmonicFieldTest/B0z_G10.out.root')
# f = root_open('BFCut.out.root')

x  = np.empty((0,1), float)
y  = np.empty((0,1), float)
z  = np.empty((0,1), float)
Bx = np.empty((0,1), float)
By = np.empty((0,1), float)
Bz = np.empty((0,1), float)
dBxdx = np.empty((0,1), float)
dBydy = np.empty((0,1), float)
dBzdz = np.empty((0,1), float)

# canvas = Canvas(width=700, height=500)
# hsurf = Hist2D(10,0,1,1000,0,1);
for evt in f.mytree:
    x  = np.append(x, evt.x)
    y  = np.append(y, evt.y)
    z  = np.append(z, evt.z)
    Bx = np.append(Bx, evt.Bx)
    By = np.append(By, evt.By)
    Bz = np.append(Bz, evt.Bz)
    dBxdx = np.append(dBxdx, evt.dBxdx)
    dBydy = np.append(dBydy, evt.dBydy)
    dBzdz = np.append(dBzdz, evt.dBzdz)
#     print(evt.Bz)
# hsurf.Draw("SURF");
# canvas.Draw()

fig, ax = plt.subplots()
ax = plt.axes(projection='3d')
ax.scatter(x, y, Bz, c=Bz, cmap='viridis', linewidth=0.5);
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$B_z$');
fig.tight_layout()
````

Great, this is as we had hoped. 

````python
fig, ax = plt.subplots()
ax = plt.axes(projection='3d')
ax.scatter(x, y, Bx, c=Bx, cmap='viridis', linewidth=0.5);
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$B_x$');
fig.tight_layout()

fig, ax = plt.subplots()
ax = plt.axes(projection='3d')
ax.scatter(x, y, By, c=By, cmap='viridis', linewidth=0.5);
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$B_y$');
fig.tight_layout()
````

And this is also consistent with the relationships among uniform gradients, as given above. PENTrack gives us the gradients values at every point on the plane as well. 

````python
fig, ax = plt.subplots()
# ax = plt.axes(projection='3d')
# ax.scatter(x, y, dBzdz, c=dBzdz, cmap='viridis', linewidth=0.5);
ax.scatter(z, dBzdz, label = r'$\frac{\partial B_z}{\partial z}$');
ax.scatter(x, dBxdx, label = r'$\frac{\partial B_x}{\partial x}$');
ax.scatter(y, dBydy+0.01, label = r'$\frac{\partial B_y}{\partial y}$');
ax.set_xlabel(r'$x_i$')
ax.set_ylabel(r'$\partial B_{x_i}$')
# ax.set_zlabel(r'$\frac{\partial B_z}{\partial z}$');
ax.legend()
fig.tight_layout()
````

Here we can see that the gradients are correct, I added a small offset to the $\frac{\partial B_y}{\partial y}$ gradient, just for visisbility.
